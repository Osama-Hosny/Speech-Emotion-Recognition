### 2-Creating The Feature Space

a) Time Domain

```python
time_data = []

def get_time_domain_features(audio, sample_rate):
    zcr = np.mean(librosa.feature.zero_crossing_rate(audio).T, axis=0)
    energy = np.mean(librosa.feature.rms(y=audio).T, axis=0)
    return np.concatenate([zcr, energy])

for i in range(len(data)):
    time_data.append(get_time_domain_features(data[i][0], data[i][1]))
x_time = np.array(time_data)
y_time = np.array(label)
print(x_time)
```

b) Frequency Domain

```python
freq_data = []
lengths_freq = []

def get_frequency_features(audio, sample_rate):
    spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sample_rate)
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sample_rate)
    spectral_rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sample_rate)

    return np.array([
        np.mean(spectral_centroid),
        np.mean(spectral_bandwidth),
        np.mean(spectral_rolloff)
    ])

for i in range(len(data)):
    freq_features = get_frequency_features(data[i][0], data[i][1])
    freq_data.append(freq_features)

x_freq = np.array(freq_data)
y_freq = np.array(label)

print("Frequency features shape:", x_freq.shape)
print("Sample frequency features:", x_freq[0])
```

c) Mel Spectrogram

```python
mel_data = []
lengths = []
def get_mel_spectrogram(audio, sample_rate):
    mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=128)
    mel_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
    mel_db = (mel_db - np.mean(mel_db)) / np.std(mel_db)
    return mel_db

for i in range(len(data)):
    mel = get_mel_spectrogram(data[i][0], data[i][1])
    mel_data.append(mel)
    lengths.append(mel.shape[1])

max_len = max(lengths)

padded_mel_data = []
for mel in mel_data:
    if mel.shape[1] < max_len:
        pad_width = max_len - mel.shape[1]
        mel = np.pad(mel, pad_width=((0, 0), (0, pad_width)), mode='constant')
    padded_mel_data.append(mel)
x_mel = np.stack(padded_mel_data)
y_mel = np.array(label)
```

```python
def prepare_dataloader(X_train, y_train, X_val, y_val, X_test, y_test, batch_size):
    X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train, dtype=torch.long)
    X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
    y_val_tensor = torch.tensor(y_val, dtype=torch.long)
    X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test, dtype=torch.long)

    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, val_loader, test_loader
```

```python
def prepare_dataloader2d(X_train, y_train, X_val, y_val, X_test, y_test, batch_size):
    # Add channel dimension (1) for 2D CNN input
    X_train_tensor = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1)  # (N, 1, 128, 216)
    X_val_tensor = torch.tensor(X_val, dtype=torch.float32).unsqueeze(1)
    X_test_tensor = torch.tensor(X_test, dtype=torch.float32).unsqueeze(1)

    y_train_tensor = torch.tensor(y_train, dtype=torch.long)
    y_val_tensor = torch.tensor(y_val, dtype=torch.long)
    y_test_tensor = torch.tensor(y_test, dtype=torch.long)

    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, val_loader, test_loader
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Set random seeds for reproducibility
np.random.seed(42)
torch.manual_seed(42)

## Model Architectures
class CNN1D(nn.Module):
    def __init__(self, input_shape, num_classes):
        super(CNN1D, self).__init__()
        self.input_shape = input_shape
        self.num_classes = num_classes
        self.conv_layers = nn.Sequential(
        nn.Conv1d(input_shape[0], 64, kernel_size=3, padding=1),
        nn.BatchNorm1d(64),
        nn.ReLU(),
        nn.MaxPool1d(2),
        nn.Dropout(0.3),

        nn.Conv1d(64, 128, kernel_size=3, padding=1),
        nn.BatchNorm1d(128),
        nn.ReLU(),
        nn.MaxPool1d(2),
        nn.Dropout(0.3),

        nn.Conv1d(128, 256, kernel_size=3, padding=1),
        nn.BatchNorm1d(256),
        nn.ReLU(),
        nn.MaxPool1d(2),
        nn.Dropout(0.3),

        nn.AdaptiveAvgPool1d(1)
        )

        self.fc_layers = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, num_classes)
         )


    def forward(self, x):
        x = self.conv_layers(x)
        x = x.squeeze(-1)  # Remove the last dimension
        x = self.fc_layers(x)
        return x

class CNN2D(nn.Module):
    """Creates a 2D CNN model for mel spectrogram features"""
    def __init__(self, input_shape, num_classes):
        super(CNN2D, self).__init__()
        self.input_shape = input_shape
        self.num_classes = num_classes
        self.conv_layers = nn.Sequential(
            # First Conv2D block
            nn.Conv2d(1, 32, kernel_size=3),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25),

            # Second Conv2D block
            nn.Conv2d(32, 64, kernel_size=3),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25),

            # Third Conv2D block
            nn.Conv2d(64, 128, kernel_size=3),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25)
        )

        # Calculate the size of the flattened features
        with torch.no_grad():
            dummy_input = torch.zeros(1, 1, *input_shape)  # (1, 1, 128, 216)
            dummy_output = self.conv_layers(dummy_input)
            flattened_size = dummy_output.view(1, -1).shape[1]

        self.fc_layers = nn.Sequential(
            nn.Linear(flattened_size, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = self.fc_layers(x)
        return x




## Main Workflow
# def main():
    # Example usage - you'll need to replace with your actual data loading code

    # 1. Load and preprocess your data
    # X_train, y_train, X_val, y_val, X_test, y_test = load_crema_dataset()
    # num_classes = len(np.unique(y_train))

    # 2. Create DataLoaders
    # batch_size = 32
    # train_loader, val_loader, test_loader = prepare_dataloader(
    #     X_train, y_train, X_val, y_val, X_test, y_test, batch_size)

    # 3. Create model (example for 1D CNN)
    # input_shape = (X_train.shape[1], X_train.shape[2])  # For 1D CNN: (channels, length)
    # model = CNN1D(input_shape, num_classes)

    # 4. Train model
    # device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    # history, best_model = train_model(model, train_loader, val_loader, epochs=50, device=device)

    # 5. Evaluate
    # accuracy, f1, cm = evaluate_model(best_model, test_loader, device)
    # print(f"Test Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}")

    # 6. Plot results
    # plot_training_history(history)
    # plot_confusion_matrix(cm, class_names=['Angry', 'Happy', 'Neutral', 'Sad'])

#     pass

# if __name__ == "__main__":
#     main()
```

```python

## Custom Training Functions
def train_epoch(model, train_loader, loss_fn, optimizer, device):

    model.train()
    total_loss = 0
    correct_preds = 0
    total_samples = 0

    for batch_idx, (x_batch, y_batch) in enumerate(train_loader):
        x_batch = x_batch.to(device)
        y_batch = y_batch.to(device)

        # Forward pass
        optimizer.zero_grad()
        predictions = model(x_batch)
        loss = loss_fn(predictions, y_batch)

        # Backward pass
        loss.backward()
        optimizer.step()

        # Calculate metrics
        total_loss += loss.item()
        _, predicted = torch.max(predictions.data, 1)
        correct_preds += (predicted == y_batch).sum().item()
        total_samples += y_batch.size(0)

    epoch_loss = total_loss / len(train_loader)
    epoch_acc = correct_preds / total_samples
    return epoch_loss, epoch_acc

def validate_epoch(model, val_loader, loss_fn, device):
    model.eval()
    total_loss = 0
    correct_preds = 0
    total_samples = 0

    with torch.no_grad():
        for batch_idx, (x_batch, y_batch) in enumerate(val_loader):
            x_batch = x_batch.to(device)
            y_batch = y_batch.to(device)

            # Forward pass
            predictions = model(x_batch)
            loss = loss_fn(predictions, y_batch)

            # Calculate metrics
            total_loss += loss.item()
            _, predicted = torch.max(predictions.data, 1)
            correct_preds += (predicted == y_batch).sum().item()
            total_samples += y_batch.size(0)

    epoch_loss = total_loss / len(val_loader)
    epoch_acc = correct_preds / total_samples
    return epoch_loss, epoch_acc

def train_model(model, train_loader, val_loader, epochs=50, learning_rate=0.001, patience=5, device='cpu'):
    model = model.to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    loss_fn = nn.CrossEntropyLoss()

    best_val_loss = float('inf')
    wait = 0
    best_model = None

    history = {
        'train_loss': [],
        'train_acc': [],
        'val_loss': [],
        'val_acc': []
    }

    # Training loop
    for epoch in range(epochs):
        print(f"\nEpoch {epoch + 1}/{epochs}")

        # Train
        train_loss, train_acc = train_epoch(model, train_loader, loss_fn, optimizer, device)
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)

        # Validate
        val_loss, val_acc = validate_epoch(model, val_loader, loss_fn, device)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)

        print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
        print(f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")

        # Early stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            wait = 0
            best_model = type(model)(model.input_shape, model.num_classes).to(device)
            best_model.load_state_dict(model.state_dict())
        # else:
        #     wait += 1
        #     if wait >= patience:
        #         print(f"\nEarly stopping at epoch {epoch + 1}")
        #         break

    return history, best_model
```

```python
def evaluate_model(model, test_loader, device):
    model.eval()
    y_true = []
    y_pred = []

    with torch.no_grad():
        for x_batch, y_batch in test_loader:
            x_batch = x_batch.to(device)
            outputs = model(x_batch)
            _, predicted = torch.max(outputs.data, 1)

            y_true.extend(y_batch.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())

    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')
    cm = confusion_matrix(y_true, y_pred)

    return accuracy, f1, cm

def plot_training_history(history):
    """Plots training and validation metrics"""
    plt.figure(figsize=(12, 4))

    # Plot loss
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Train Accuracy')
    plt.plot(history['val_acc'], label='Val Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(cm, class_names):
    """Plots a confusion matrix"""
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```
